```python
# Task 2 Sadhav Singh
# ID:CT6WDS79

# Step 1: Install the required libraries
!pip install diffusers --upgrade
!pip install invisible_watermark transformers accelerate safetensors gradio

# Step 2: Import necessary libraries
from diffusers import DiffusionPipeline
import torch
import gradio as gr
import matplotlib.pyplot as plt

# Step 3: Initialize the base pipeline
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
)

pipe.to("cuda")

# Step 4: Define the function to generate images
def generate_image(prompt, negative_prompt):
    images = pipe(prompt=prompt, negative_prompt=negative_prompt).images[0]
    return images

# Step 5: Create a Gradio interface
interface = gr.Interface(
    fn=generate_image,
    inputs=[
        gr.Textbox(label="Prompt", value="A vibrant sunset over the city skyline with assassin's creed buildings."),
        gr.Textbox(label="Negative Prompt", value="Avoid including any water elements in the scene.")
    ],
    outputs=gr.Image(),
    title="Stable Diffusion XL",
    description="Generate images using Stable Diffusion XL"
)

# Step 6: Launch the Gradio app
interface.launch()
```
